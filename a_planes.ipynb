```python
# 2) Plot a line: y = a x + b
import numpy as np
import matplotlib.pyplot as plt

# Coefficients
a, b = 2.0, -1.0

x = np.linspace(-10, 10, 400)
y = a*x + b

plt.figure()
plt.plot(x, y)
plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'y = {a}x {b}')
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.show()
```

```python
# 1) Plot a 3D plane: Z = A x + B y + C
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 (needed for 3D)

# Coefficients
A, B, C = 3.0, -20.5, 2.0

# Grid for x and y
x = np.linspace(-5, 5, 60)
y = np.linspace(-5, 5, 60)
X, Y = np.meshgrid(x, y)
Z = A*X + B*Y + C

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, rstride=1, cstride=1, linewidth=0, antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(f'Plane: z = {A}x + {B}y + {C}')

plt.show()
```

```python
# 3) Plot a quadratic: y = a x^2 + b x + c
import numpy as np
import matplotlib.pyplot as plt

# Coefficients
a, b, c = 1.0, -3.0, 2.0

x = np.linspace(-10, 10, 400)
y = a*x**2 + b*x + c

plt.figure()
plt.plot(x, y)
plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'y = {a}x + {b}x + {c}')
plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.show()
```
